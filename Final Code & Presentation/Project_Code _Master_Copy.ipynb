```python
#Import Modules 
import pandas as pd
import datetime as dt


bitcoin_csv = "bitcoin_data.csv"

df = pd.read_csv(bitcoin_csv)
df
```

```python
#Isolate Columns [Times etc...]
df["Timestamp"] = pd.to_datetime(df['Timestamp'], unit= 's')
df_hour = df[df["Timestamp"].dt.minute==0]
df_hour.head()
```

```python
df_hour
```

```python
df_hour.tail()
```

```python
df_hour.drop(['Weighted_Price'], axis=1)
df_hour_clean = df_hour.drop(['Weighted_Price'], axis=1)
df_hour_clean.head()
```

```python
df_hour_clean.to_csv('cleaned_bitcoin_compressed.csv', index = False)
```

```python
weekends = [4,5,6,0]

df_weekend_clean =df_hour_clean[df_hour_clean['Timestamp'].dt.weekday.isin(weekends)]
df_weekend_clean
df_weekend_clean.info()
```

```python
df_weekend_clean
```

```python
#NOTE

year= [2021]
month = [1,2,3]

df_feb_2021 =df_weekend_clean[(df_weekend_clean['Timestamp'].dt.year.isin(year)) & (df_weekend_clean['Timestamp'].dt.month.isin(month))]
df_feb_2021.info()
```

```python
df_feb_2021.head()
```

```python
# For Loop to extract trade Data
from tqdm import tqdm
import time
df_feb_2021

results = []
for idx1, row1 in tqdm(df_feb_2021.iterrows(), total=len(df_feb_2021)):
    for idx2, row2 in df_feb_2021.iterrows():
        if idx2>idx1:
            trade_open =row1["Timestamp"]
            trade_close = row2["Timestamp"]
            enter_price = row1["Close"] 
            exit_price = row2["Close"]
            results.append([trade_open, trade_close, enter_price, exit_price]) 

# print(results)
results_df = pd.DataFrame(results) 
results_df.head()

```

```python
results_df.rename(columns={0:"TRADE_TIME_ON", 1:"TRADE_TIME_OFF",2:"ENTER_PRICE", 3:"EXIT_PRICE"})
results_df = results_df.rename(columns={0:"TRADE_TIME_ON", 1:"TRADE_TIME_OFF",2:"ENTER_PRICE", 3:"EXIT_PRICE"})
results_df
```

```python
#Short trade caluclation
results_df['P&L'] = results_df['ENTER_PRICE'] - results_df['EXIT_PRICE']
results_df.head()
```

```python
results_df
```

```python
results_df.to_csv('Jan_Feb_March_2021_P&L.csv', index = False)
```

```python
# Regression v Time Plot

from scipy.stats import linregress
import scipy.stats as st

volume_adjusted_macro = "volume_macro_btc.csv"

df = pd.read_csv(volume_adjusted_macro)
df
```

```python
x_values = df['Adjusted']
y_values = df['Close']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(20,40),fontsize=15,color="red")
plt.xlabel('Adjusted Volime (millions $USD)')
plt.ylabel('Price $USD')
plt.show()
```

```python
correlation = (st.pearsonr(df['Adjusted'], df['Close'])[0])
print(f"The correlation between price and volume is {correlation}")
```
