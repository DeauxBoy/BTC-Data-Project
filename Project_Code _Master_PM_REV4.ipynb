```python
#df_weekend_clean
from tqdm import tqdm
import time
df_feb_2021

results = []
for idx1, row1 in tqdm(df_feb_2021.iterrows(), total=len(df_feb_2021)):
    for idx2, row2 in df_feb_2021.iterrows():
        if idx2>idx1:
            trade_open =row1["Timestamp"]
            trade_close = row2["Timestamp"]
            enter_price = row1["Close"] 
            exit_price = row2["Close"]
            results.append([trade_open, trade_close, enter_price, exit_price]) 

#print(results)
results_df = pd.DataFrame(results) 
results_df.head()
#pd.DataFrame({ "TRADE ON" : results, "TRADE OFF": results,"ENTER PRICE" : results, "EXIT PRICE": results})


#results.to_excel("btc_weekend.xlsx") 
        

    #print(idx1 , row1)
    #time.sleep(0.01)

```
